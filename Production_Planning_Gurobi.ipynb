```python
#Step 1: After Relaxation
from gurobipy import*

# create empty model
m = Model("Example2")

# add variables
x = m.addVars(range(1,3), vtype=GRB.CONTINUOUS, name = "Vars")

# set objective function
m.setObjective(2 * x[1] + 3 * x[2], GRB.MAXIMIZE)

# add constraints
m.addConstr(x[1] + 2*x[2] <= 10)
m.addConstr(3 * x[1] + 4*x[2] <= 25)
#m.addConstr(sum(x[i] for i in range(2)) >= 1)  # here we use sum function of Python, whose argument is a list or generator expression

# solve model
m.optimize()

# print the optimal objective value
print('\n Optimal Objective value is equal to ',m.objVal)

# print the optimal solution
m.printAttr('X')
print("Opt. obj. value is %0.2f" %float(m.objVal))
for i in range(1,3):
    print("The X[%d] values are " %i, "%0.2f" %(m.x[i-1]))
```
